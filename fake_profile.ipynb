```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['profile pic']=le.fit_transform(df['profile pic'])
df['nums/length username']=le.fit_transform(df['nums/length username'])
df['fullname words']=le.fit_transform(df['fullname words'])
df['nums/length fullname']=le.fit_transform(df['nums/length fullname'])
df['name==username']=le.fit_transform(df['name==username'])
df['description length']=le.fit_transform(df['description length'])
df['external URL']=le.fit_transform(df['external URL'])
df['private']=le.fit_transform(df['private'])
df['posts']=le.fit_transform(df['posts'])
df['followers']=le.fit_transform(df['followers'])
df['follows']=le.fit_transform(df['follows'])
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
x = df.drop(['fake'], axis=1)
y = df['fake']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.3, random_state = 42)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
y_pred = dt.predict(x_test)
acc = accuracy_score(y_pred,y_test)
acc
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
y_pred = rf.predict(x_test)
acc = accuracy_score(y_pred,y_test)
acc
```

```python
abc = [[1,0,3,0,0,0,0,1,112,176,151]]
result = rf.predict(abc)
result[0]
```
